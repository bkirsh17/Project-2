```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.io.ascii 
import astropy.constants as ac
import astropy.units as u
import pandas as pd
```

```python
exoplanets = pd.read_csv("PS_2023.02.21_11.18.39.csv", comment = '#')
```

```python
exoplanets.columns
```

```python
plt.plot(exoplanets["pl_bmasse"], exoplanets["pl_rade"], "c.",label = "NEA Data")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Mass [M$_{Earth}$]")
plt.ylabel("Radius [R$_{Earth}$]")
plt.plot(ac.M_jup/ac.M_earth, ac.R_jup/ac.R_earth, "bP", markersize=8, label = "Jupiter")
plt.plot(359.672, 13.198, "r*", markersize = 14, label = "HD 189733 b")
plt.legend()
plt.title('Radius vs. Mass', fontsize = 17)
plt.savefig("radius_mass.png")
```

```python
exoplanets["pl_mass"] = exoplanets["pl_bmasse"]*(5.972E+24)
exoplanets["pl_rad"] = exoplanets["pl_rade"]*(6.38E+6)
```

```python
exoplanets["density"] = exoplanets["pl_mass"]/((4/3)*np.pi*(exoplanets["pl_rad"])**3)
```

```python
plt.plot(exoplanets["density"], exoplanets["pl_rade"], "c.",label = "NEA Data")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Density [kg m$^{-3}$]")
plt.ylabel("Radius [R$_{Earth}$]")
plt.plot(ac.M_jup/((4/3)*np.pi*(ac.R_jup)**3), ac.R_jup/ac.R_earth, "bP", markersize=8, label = "Jupiter")
plt.plot(359.672*(5.972E+24)/((4/3)*np.pi*(13.198*(6.38E+6))**3), 13.198, "r*", markersize = 14, label = "HD 189733 b")
plt.legend(loc= "upper right")
plt.title('Radius vs. Density', fontsize = 17)
plt.savefig("radius_density.png")
```

```python
plt.plot(exoplanets["density"], exoplanets["pl_bmasse"], "c.",label = "NEA Data")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Density [kg m$^{-3}$]")
plt.ylabel("Mass [M$_{Earth}$]")
plt.plot(ac.M_jup/((4/3)*np.pi*(ac.R_jup)**3), ac.M_jup/ac.M_earth, "bP", markersize=8, label = "Jupiter")
plt.plot(359.672*(5.972E+24)/((4/3)*np.pi*(13.198*(6.38E+6))**3), 359.672, "r*", markersize = 14, label = "HD 189733 b")
plt.legend(loc = "upper right")
plt.title('Mass vs. Density', fontsize = 17)
plt.savefig("mass_density.png")
```
